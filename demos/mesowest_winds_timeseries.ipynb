```python
%matplotlib inline
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import sys

sys.path.append('/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v2')

from BB_MesoWest.MesoWest_timeseries import get_mesowest_ts
from BB_wx_calcs.wind import *

# Hour locator
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
formatter = DateFormatter('%b-%d\n%H:%M')

import matplotlib as mpl 
mpl.rcParams['figure.figsize'] = [10, 5]
mpl.rcParams['figure.titlesize'] = 12
mpl.rcParams['figure.titleweight'] = 'bold'
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['axes.titlesize'] = 14
mpl.rcParams['lines.linewidth'] = 1.8
mpl.rcParams['grid.linewidth'] = .25
mpl.rcParams['figure.subplot.wspace'] = 0.05
mpl.rcParams['figure.subplot.hspace'] = 0.05
mpl.rcParams['legend.fontsize'] = 10
mpl.rcParams['legend.framealpha'] = .75
mpl.rcParams['legend.loc'] = 'best'
mpl.rcParams['legend.scatterpoints'] = 1
mpl.rcParams['savefig.bbox'] = 'tight'
```

```python
sDATE = datetime(2017, 9, 30, 0)
eDATE = datetime(2017, 10, 1, 6)

#DATE = datetime(2018, 4, 16, 9)
#DATE = datetime(2018, 4, 17, 9)

#sDATE = datetime(2018, 4, 12, 0)
#eDATE = datetime(2018, 4, 13, 0)


a = get_mesowest_ts('UKBKB', sDATE, eDATE, variables='wind_speed,wind_direction,wind_gust')
b = get_mesowest_ts('C9348', sDATE, eDATE, variables='wind_speed,wind_direction,wind_gust')
c = get_mesowest_ts('UCC13', sDATE, eDATE, variables='wind_speed,wind_direction,wind_gust')
c1 = get_mesowest_ts('QSF', sDATE, eDATE, variables='wind_speed,wind_direction,wind_gust')
```

```python
A = a

local = [d - timedelta(hours=6) for d in A['DATETIME']]

fig = plt.figure()
ax1 = fig.add_subplot(111)

u, v = wind_spddir_to_uv(A['wind_speed']*2.2369, A['wind_direction'])
idx = mpl.dates.date2num(local)

Q = plt.quiver(idx, 30, u, v, label='Direction', width=.003,
               color=[.7,.7,.7])

ax = plt.gca()
ax.xaxis.set_major_locator(mdates.HourLocator(range(0, 24, 3)))
ax.xaxis.set_minor_locator(mdates.HourLocator(range(0, 24, 1)))
ax.xaxis.set_major_formatter(formatter)


plt.sca(ax1)
plt.plot(local, A['wind_speed']*2.2369,
         label='Wind Speed',
         color='darkblue',
         zorder=1)
plt.scatter(local, A['wind_gust']*2.2369,
            color='darkgreen',
            label='Gust')
ax1.set_xlim([sDATE- timedelta(hours=6), eDATE- timedelta(hours=6)])
ax1.set_ylim(ymin=0, ymax=60)
ax1.set_yticks(range(0,61,10))
ax1.set_ylabel(r'Wind Speed (MPH)')

legend = ax1.legend(frameon=True, framealpha=.9, loc='upper left', scatterpoints=1)
legend.get_frame().set_linewidth(0)
legend.set_zorder(100)

ax1.xaxis.set_major_locator(mdates.HourLocator(range(0, 24, 3)))
ax1.xaxis.set_minor_locator(mdates.HourLocator(range(0, 24, 1)))
ax1.xaxis.set_major_formatter(formatter)

ax1.grid()
plt.title('%s %s' % (A['STID'], A['NAME']), fontweight='semibold')
plt.savefig('./%s_wind' % A['STID'])
```
