# Plot a MesoWest timeseries

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = [16,9]
mpl.rcParams['xtick.labelsize'] = 15
mpl.rcParams['ytick.labelsize'] = 15
mpl.rcParams['axes.labelsize'] = 17
mpl.rcParams['axes.titlesize'] =20
mpl.rcParams['grid.linewidth'] = .25
mpl.rcParams['legend.fontsize'] = 15
mpl.rcParams['legend.loc'] = 'best'
mpl.rcParams['savefig.bbox'] = 'tight'

from matplotlib.dates import DateFormatter, HourLocator, MonthLocator
from datetime import datetime, timedelta
from collections import OrderedDict

import numpy as np

import sys
sys.path.append('/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v2')
from BB_MesoWest.MesoWest_timeseries import get_mesowest_ts
from BB_wx_calcs.wind import wind_spddir_to_uv
```

```python
stn = 'UKBKB'
sDATE = datetime(2013, 1, 1)
eDATE = datetime(2019, 1, 1)
variable = 'air_temp'

a = get_mesowest_ts(stn, sDATE, eDATE, variable)
```

```python
plt.plot(a['DATETIME'], a[variable]*9/5.+32)
plt.title(stn)
plt.ylabel('Air Temperature (F)')

plt.xlim([sDATE, eDATE])
plt.ylim([-5,105])

plt.grid()

plt.gca().xaxis.set_major_locator(MonthLocator(bymonth=range(1,13,13)))
dateFmt = DateFormatter('%b\n%Y')
plt.gca().xaxis.set_major_formatter(dateFmt)

plt.savefig('%s' % sDATE.year)
```
